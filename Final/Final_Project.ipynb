```python
all_df = pd.read_csv('./Data/us_etf_price.csv', parse_dates=['Date'])
```

```python
df = all_df.pivot(columns='tic', values='Adj Close', index='Date')
df = df.loc['2011-10-28':]
#df = df[ticker]
df.tail()
```

```python
print('df Shape before cleaning =', df.shape)

missing_percentage = df.isnull().mean().sort_values(ascending=False)
missing_percentage.head(10)
dropped_list = sorted(list(missing_percentage[missing_percentage > 0.2].index))
df.drop(labels=dropped_list, axis=1, inplace=True)

print('df Shape after cleaning =', df.shape)

df = df.fillna(method='ffill')

df = df.dropna(axis=1)
df = np.log(df).diff().dropna()
```

## 길이 나누기
