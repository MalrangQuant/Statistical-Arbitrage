## Set Return

$$R_{i, t+\Delta t} = \frac{S_{i,(t+\Delta t)}-S_{i,t}}{S_{i,t}}$$

```python
d_ret_df = raw_data.pct_change().dropna()
d_ret_df.head()
```

```python
'''Standard Return df'''
df = (d_ret_df - d_ret_df.mean()) / d_ret_df.std()
df.head()
```

```python
df['AAPL'] @ df['AMZN'].T / len(df['AAPL'])
```

```python
cor = pd.DataFrame(index=df.columns, columns=df.columns)
cor['AAPL'].loc['AAPL']
```

```python
from itertools import combinations as cb

cor = pd.DataFrame(index=df.columns, columns=df.columns)

for x in cb(df.columns, 2):

    checker = df[list(x)]
    rho = (checker.iloc[:,0] @ checker.iloc[:,1].T) / len(df)
    cor[x[0]].loc[x[1]] = rho
    cor[x[1]].loc[x[0]] = rho

cor = cor.fillna(1).astype(float)
cor
```
